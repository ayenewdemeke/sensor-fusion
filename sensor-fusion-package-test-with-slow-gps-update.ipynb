```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sfusion import kalman as km
```

## Simulated Data

```python
data = pd.read_csv('missing_gps_data.csv')
```

## 1. LinearKFGPSAccelerometer1D

### 1.1. Raw Data Plot

```python
# Filter out rows where GPS_X is NaN
gps_data = data.dropna(subset=['GPS_X'])

# Plot the filtered GPS X data as points
plt.figure(figsize=(10, 6))
plt.scatter(gps_data['Time'], gps_data['GPS_X'], label='GPS X')
plt.xlabel('Time [s]')
plt.ylabel('X Position')
plt.title('Filtered GPS X Position Data')
plt.legend()
plt.show()

# Plot the raw accelerometer X data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Accel_X'], label='Raw Accelerometer X', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Acceleration X')
plt.title('Raw Accelerometer X Data')
plt.legend()
plt.show()
```

### 1.2. Sensor Fusion

```python
# Initialize the Kalman filter for the x direction
initial_state = [data['GPS_X'].dropna().iloc[0], 0]  # initial position and velocity
initial_covariance = np.eye(2)
process_noise = np.eye(2) * 0.01
measurement_noise = np.eye(1) * 0.1

kf = km.LinearKFGPSAccelerometer1D(initial_state, initial_covariance, process_noise, measurement_noise)

# Run the Kalman filter
fused_positions = []
fused_velocities = []

dt = 0.05  # Fixed time step of 50 milliseconds

for i in range(len(data)):
    acceleration = data['Accel_X'].iloc[i]
    kf.predict(dt=dt, acceleration=acceleration)
    
    if not np.isnan(data['GPS_X'].iloc[i]):
        kf.update(measurement=data['GPS_X'].iloc[i])
    
    fused_positions.append(kf.state[0])
    fused_velocities.append(kf.state[1])

fused_positions = np.array(fused_positions)
fused_velocities = np.array(fused_velocities)
```

### 1.3. Fused Data Plot

```python
# Plot the fused trajectory data along with the raw GPS data
plt.figure(figsize=(10, 6))
plt.scatter(gps_data['GPS_X'], gps_data['GPS_Y'], label='Raw GPS')
plt.plot(fused_positions_x, fused_positions_y, label='Fused', linestyle='-', color='#ff7f0e')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Raw GPS and Fused Data')
plt.legend()
plt.show()

# Plot the fused velocity data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], fused_velocities_x, label='Fused X Velocity', linestyle='-')
plt.plot(data['Time'], fused_velocities_y, label='Fused Y Velocity', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Velocity')
plt.title('Fused Velocity Data')
plt.legend()
plt.show()
```

## 2. LinearKFGPSAccelerometer2D

### 2.1. Raw Data Plot

```python
# Plot the raw GPS X and Y data
plt.figure(figsize=(10, 6))
plt.scatter(data['Time'], data['GPS_X'], label='Raw GPS X')
plt.scatter(data['Time'], data['GPS_Y'], label='Raw GPS Y')
plt.xlabel('Time [s]')
plt.ylabel('Position')
plt.title('Raw GPS Position Data')
plt.legend()
plt.show()

# Plot the raw accelerometer X and Y data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Accel_X'], label='Raw Accelerometer X', linestyle='-')
plt.plot(data['Time'], data['Accel_Y'], label='Raw Accelerometer Y', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Acceleration')
plt.title('Raw Accelerometer Data')
plt.legend()
plt.show()
```

### 2.2. Sensor Fusion

```python
# Initialize the Kalman filter for the x and y directions
initial_state = [data['GPS_X'].dropna().iloc[0], data['GPS_Y'].dropna().iloc[0], 0, 0]  # initial positions and velocities
initial_covariance = np.eye(4)
process_noise = np.eye(4) * 0.01
measurement_noise = np.eye(2) * 0.1

kf = km.LinearKFGPSAccelerometer2D(initial_state, initial_covariance, process_noise, measurement_noise)

# Run the Kalman filter
fused_positions = []
fused_velocities = []

dt = 0.05  # Fixed time step of 50 milliseconds

for i in range(len(data)):
    acceleration = [data['Accel_X'].iloc[i], data['Accel_Y'].iloc[i]]
    kf.predict(dt=dt, acceleration=acceleration)
    
    if not np.isnan(data['GPS_X'].iloc[i]) and not np.isnan(data['GPS_Y'].iloc[i]):
        measurement = [data['GPS_X'].iloc[i], data['GPS_Y'].iloc[i]]
        kf.update(measurement=measurement)
    
    fused_positions.append(kf.state[:2])
    fused_velocities.append(kf.state[2:])

fused_positions = np.array(fused_positions)
fused_velocities = np.array(fused_velocities)
```

### 2.3. Fused Data Plot

```python
# Plot the fused trajectory data along with the raw GPS data
plt.figure(figsize=(10, 6))
plt.scatter(gps_data['GPS_X'], gps_data['GPS_Y'], label='Raw GPS', marker='o')
plt.plot(fused_positions[:, 0], fused_positions[:, 1], label='Fused', linestyle='-', color='#ff7f0e')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Raw GPS and Fused Data')
plt.legend()
plt.show()

# Plot the fused velocity data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], fused_velocities[:, 0], label='Fused X Velocity', linestyle='-')
plt.plot(data['Time'], fused_velocities[:, 1], label='Fused Y Velocity', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Velocity')
plt.title('Fused Velocity Data')
plt.legend()
plt.show()
```

### 2.4. Zoomed-in Plot

```python
# Define the range you want to zoom in on (adjust indices based on your data)
start_time = 10
end_time = 30

# Convert times to the index range for the zoomed-in plot
start_index = data[data['Time'] >= start_time].index[0]
end_index = data[data['Time'] <= end_time].index[-1]

# Plot the fused data with zoom for the trajectory
plt.figure(figsize=(10, 6))
plt.scatter(data['GPS_X'][start_index:end_index], data['GPS_Y'][start_index:end_index], label='Raw GPS Trajectory')
plt.plot(fused_positions[start_index:end_index, 0], fused_positions[start_index:end_index, 1], label='Fused Trajectory', linestyle='-', color='#ff7f0e')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Raw GPS Trajectory and Fused Trajectory Data (Zoomed In)')
plt.legend()
plt.show()
```

## 3. LinearKFGPSAccelerometerGyro2D

### 3.1. Raw Data Plot

```python
# Plot the raw accelerometer X and Y data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Accel_X'], label='Raw Accelerometer X', linestyle='-')
plt.plot(data['Time'], data['Accel_Y'], label='Raw Accelerometer Y', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Acceleration')
plt.title('Raw Accelerometer Data')
plt.legend()
plt.show()

# Plot the raw gyroscope data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Gyro_Z'], label='Raw Gyroscope Z', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Angular Velocity Z')
plt.title('Raw Gyroscope Data')
plt.legend()
plt.show()
```

### 3.2. Sensor Fusion

```python
# Initialize the Kalman filter for the x and y directions
initial_state = [data['GPS_X'].dropna().iloc[0], data['GPS_Y'].dropna().iloc[0], 0, 0, 0]  # initial positions, velocities, and orientation
initial_covariance = np.eye(5)
process_noise = np.eye(5) * 0.01
measurement_noise = np.eye(2) * 0.1

kf = km.LinearKFGPSAccelerometerGyro2D(initial_state, initial_covariance, process_noise, measurement_noise)

# Run the Kalman filter
fused_positions = []
fused_velocities = []
orientations = []

dt = 0.05  # Fixed time step of 50 milliseconds

for i in range(len(data)):
    acceleration = [data['Accel_X'].iloc[i], data['Accel_Y'].iloc[i]]
    angular_velocity = data['Gyro_Z'].iloc[i]
    kf.predict(dt=dt, control_input=[acceleration[0], acceleration[1], angular_velocity])
    
    if not np.isnan(data['GPS_X'].iloc[i]) and not np.isnan(data['GPS_Y'].iloc[i]):
        kf.update(measurement=[data['GPS_X'].iloc[i], data['GPS_Y'].iloc[i]])
    
    fused_positions.append(kf.state[:2])
    fused_velocities.append(kf.state[2:4])
    orientations.append(kf.state[4])

fused_positions = np.array(fused_positions)
fused_velocities = np.array(fused_velocities)
orientations = np.array(orientations)
```

### 3.3. Fused Data Plot

```python
# Plot the fused trajectory data along with the raw GPS data
plt.figure(figsize=(10, 6))
plt.scatter(data['GPS_X'], data['GPS_Y'], label='Raw GPS')
plt.plot(fused_positions[:, 0], fused_positions[:, 1], label='Fused', linestyle='-', color='#ff7f0e')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Raw GPS and Fused Data')
plt.legend()
plt.show()

# Plot the fused velocity data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], fused_velocities[:, 0], label='Fused X Velocity', linestyle='-')
plt.plot(data['Time'], fused_velocities[:, 1], label='Fused Y Velocity', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Velocity')
plt.title('Fused Velocity Data')
plt.legend()
plt.show()

# Convert orientation from radians to degrees
orientations_degrees = np.degrees(orientations)

# Plot the fused orientation data in degrees
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], orientations_degrees, label='Fused Orientation', linestyle='-')
plt.xlabel('Time [s]')
plt.ylabel('Orientation (degrees)')
plt.title('Fused Orientation Data')
plt.legend()
plt.show()
```

### 3.4. Zoomed-in Plot

```python
# Define the range you want to zoom in on (adjust indices based on your data)
start_time = 10
end_time = 50

# Convert times to the index range for the zoomed-in plot
start_index = data[data['Time'] >= start_time].index[0]
end_index = data[data['Time'] <= end_time].index[-1]

# Plot the fused data with zoom for the trajectory
plt.figure(figsize=(10, 6))
plt.scatter(data['GPS_X'][start_index:end_index], data['GPS_Y'][start_index:end_index], label='Raw GPS Trajectory')
plt.plot(fused_positions[start_index:end_index, 0], fused_positions[start_index:end_index, 1], label='Fused Trajectory', linestyle='-', color='#ff7f0e')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Raw GPS Trajectory and Fused Trajectory Data (Zoomed In)')
plt.legend()
plt.show()
```
